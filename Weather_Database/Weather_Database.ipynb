```python
# import dependencies & API key
import pandas as pd
import numpy as np
import requests
import time
from config import weather_api_key
```

```python
# latitudes and longitudes
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lng = zip(lats, lngs)
lat_lng
```

```python
# adding latitudes and longitudes to a list
coordinates = list(lat_lng)
```

```python
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy
```

```python
# Create a list for holding the cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
```

```python
#base url for API call
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# create empty list to hold city weather API data
city_data = []

# print a beginning of the logging process
print("Beginning Data Retrieval     ")
print("-----------------------------")

# initialize counters
record_count = 1
set_count = 1

# initialize For Loop for all cities in the 'cities' list
for i, city in enumerate(cities):
    
    # grouping cities in sets of 50 so we don't break the API key
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(60)
        
    # URL per city
    city_url = base_url + "&q=" + city.replace(" ","+")
    
    # logging the process
    print(f"Processing Record {record_count} of set {set_count} | {city}")
    
    # increase record_count by 1
    record_count += 1
    
# Run the API request for every city
    try:
        
        # API call + JSONify
        city_weather = requests.get(city_url).json()
        
        # grabbing required data...
        
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_desc = city_weather["weather"][0]["description"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]        
        city_country = city_weather["sys"]["country"]
        
        
        # add the data to our city_data list
        city_data.append({
            "City": city.title(),
            "Country": city_country,
            "Lat": city_lat,
            "Lng": city_lng,
            "Max Temp": city_max_temp,
            "Humidity": city_humidity,
            "Cloudiness": city_clouds,
            "Wind Speed": city_wind,
            "Weather Description": city_desc})
     
    # Skip city if error
    except:
        print("City not found, skipping...")
        pass
        

# Indicate data retrieval's complete
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
    
```

```python
#create DataFrame from city_data list
city_data_df = pd.DataFrame(city_data)
```
